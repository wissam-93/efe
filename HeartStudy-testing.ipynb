```python
import scipy 
from scipy.stats.stats import pearsonr
```

```python
#correlation manual 
age = df['age']
anaemia = df['anaemia']
creatinine_phosphokinase = df['creatinine_phosphokinase']
diabetes = df['diabetes']
ejection_fraction = df['ejection_fraction']
high_blood_pressure = df['high_blood_pressure']
platelets = df['platelets']
serum_creatinine = df['serum_creatinine']
serum_sodium = df['serum_sodium']
sex = df['sex']
smoking = df['smoking']
time = df['time']

DEATH_EVENT = df['DEATH_EVENT']


pearsonr_coefficent,p_value = pearsonr(sex,DEATH_EVENT)
print ('Pearson relation %0.3f'% pearsonr_coefficent)
```

```python
df = pd.DataFrame(df.drop('sex', axis=1))
```

```python
df.head()
```

```python
#Selecting Feature
feature_cols = ['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','smoking','time']

x = df[feature_cols]

y = df.DEATH_EVENT

print(x.head())
```

```python
print(y)
```

```python
#Model Development and Prediction
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
accuracy  = LogisticRegression()
accuracy.fit(x_train,y_train)
y_pred=accuracy.predict(x_test)

print(accuracy.score(x_test,y_test))
```
